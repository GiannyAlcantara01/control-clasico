### **Parte 1: Análisis del sistema base (en una celda de Python).**


```python
# Importar las bibliotecas necesarias.
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# --- Definición de la planta Gp(s). ---
num_p = [5]
den_p = [1, 6, 5]
Gp = ct.tf(num_p, den_p)

# --- Sistema en lazo cerrado con retroalimentación. ---
lazo_cerrado = ct.feedback(Gp, 1)
print("Función de Transferencia del sistema base (lazo cerrado):")
print(lazo_cerrado)

# --- Simulación de la respuesta al escalón. ---
t = np.linspace(0, 5, 501)
t, y_base = ct.step_response(lazo_cerrado, T=t)

# --- Indicar las raíces de s (Polos del sistema). ---
# Usamos la función ct.poles() para obtener las raíces del denominador.
polos = ct.poles(lazo_cerrado)

# Imprimimos el resultado.
print("\nLas raíces de s (polos) son:")
print(polos)

# --- Visualización del Mapa de Polos y Ceros ---
plt.figure(figsize=(8, 6))
ct.pzmap(lazo_cerrado, title='Mapa de Polos (sistema de sistema lazo cerrado)')
plt.grid(True)
plt.show()

# --- Graficación de la respuesta base. ---
plt.figure(figsize=(8, 6))
plt.plot(t, y_base, label="Respuesta sin controlador.")
plt.axhline(y=1, color="r", linestyle=":", label="Referencia.")

plt.title("Respuesta del sistema en lazo cerrado.")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()

# --- Verificación de las métricas de desempeño. ---
print("Métricas de desempeño del sistema base:")
info = ct.step_info(sistema_base)

# step_info devuelve un diccionario, podemos imprimir los valores clave.
print(f"  - Sobreimpulso (Overshoot): {info['Overshoot']:.2f}%")
print(f"  - Tiempo de establecimiento (SettlingTime): {info['SettlingTime']:.2f} s")

# El error en estado estacionario se verifica observando el valor final.
valor_final = y_base[-1]
print(f"  - Valor final: {valor_final:.4f}")
print(f"  - Error en estado estacionario (1 - valor_final): {1 - valor_final:.4f}")
```

### **Parte 2: Proceso de diseño del PID.**

```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# --- Definición de la planta Gp(s) (la misma del ejemplo) ---
num_p = [5]
den_p = [1, 6, 5]
Gp = ct.tf(num_p, den_p)

# --- Sistema base (sin controlador) para comparar ---
sistema_base = ct.feedback(Gp, 1)
t = np.linspace(0, 5, 501)
t_base, y_base = ct.step_response(sistema_base, T=t)

# --- Sintonización del Controlador PID ---
# Estas son las ganancias sintonizadas para cumplir los requisitos:
Kp = 12.0
Ki = 10.0
Kd = 2.0

# --- Creación del controlador PID ---
# Gc(s) = (Kd*s^2 + Kp*s + Ki) / s
num_c = [Kd, Kp, Ki]
den_c = [1, 0]
Gc_PID = ct.tf(num_c, den_c)

# --- Sistema en lazo cerrado con controlador PID ---
sistema_PID = ct.feedback(Gc_PID * Gp, 1)

# --- Simulación de la respuesta al escalón del PID ---
t_pid, y_pid = ct.step_response(sistema_PID, T=t)

# --- Verificación de las métricas de desempeño ---
print("Métricas de desempeño del sistema con PID:")
info_PID = ct.step_info(sistema_PID, SettlingTimeThreshold=0.02)

print(f"  - Sobreimpulso (Overshoot): {info_PID['Overshoot']:.2f}%")
print(f"  - Tiempo de establecimiento (SettlingTime): {info_PID['SettlingTime']:.2f} s")
valor_final = y_pid[-1]
print(f"  - Valor final: {valor_final:.4f}")
print(f"  - Error en estado estacionario (1 - valor_final): {1 - valor_final:.4f}")
```

##### **El razonamiento para las ganancias ($K_p=12.0$, $K_i=10.0$, $K_d=2.0$) es el siguiente:**

1.  **Para el error cero ($e_{ss} = 0$):** Se necesita la acción integral. Se usó $K_i = 10.0$ para asegurar que el sistema elimine cualquier error a largo plazo.
2.  **Para el tiempo ($T_s < 1.0s$):** Se requiere una respuesta veloz. La ganancia proporcional $K_p = 12.0$ actúa como el acelerador principal para que el sistema reaccione y se estabilice rápidamente.
3.  **Para el sobreimpulso bajo (< 5%):** Un $K_p$ alto (como 12.0) por sí solo causa muchas oscilaciones. Para contrarrestar esto, la ganancia derivativa $K_d = 2.0$ actúa como un freno o amortiguador que reduce el sobreimpulso y mantiene la respuesta estable.

En resumen, $K_i$ elimina el error, $K_p$ da la velocidad, y $K_d$ controla el sobreimpulso que $K_p$ genera.

### **Parte 3: Implementación del controlador PID sintonizado.**

```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# --- Definición de la planta Gp(s) (la misma del ejemplo) ---
num_p = [5]
den_p = [1, 6, 5]
Gp = ct.tf(num_p, den_p)

# --- Sistema base (sin controlador) para comparar ---
sistema_base = ct.feedback(Gp, 1)
t = np.linspace(0, 5, 501)
t_base, y_base = ct.step_response(sistema_base, T=t)

# --- Sintonización del Controlador PID ---
# Estas son las ganancias sintonizadas para cumplir los requisitos:
Kp = 12.0
Ki = 10.0
Kd = 2.0

# --- Creación del controlador PID ---
# Gc(s) = (Kd*s^2 + Kp*s + Ki) / s
num_c = [Kd, Kp, Ki]
den_c = [1, 0]
Gc_PID = ct.tf(num_c, den_c)

# --- Sistema en lazo cerrado con controlador PID ---
sistema_PID = ct.feedback(Gc_PID * Gp, 1) 
print("Función de Transferencia del sistema con PID:")
print(sistema_PID)

# --- Simulación de la respuesta al escalón del PID ---
t_pid, y_pid = ct.step_response(sistema_PID, T=t)

# --- Indicar las raíces de s (Polos del sistema). ---
# Usamos la función ct.poles() para obtener las raíces del denominador.
polos = ct.poles(sistema_PID)

# Imprimimos el resultado.
print("\nLas raíces de s (polos) son:")
print(polos)

# --- Visualización del Mapa de Polos y Ceros ---
plt.figure(figsize=(8, 6))
ct.pzmap(sistema_PID, title='Mapa de Polos y Ceros (sistema de sistema lazo cerrado)')
plt.grid(True)
plt.show()

# --- Graficación de la comparativa ---
plt.figure(figsize=(8, 6))
plt.plot(t_base, y_base, label="Respuesta sin controlador (Base).", linestyle='--')
plt.plot(t_pid, y_pid, label=f"Controlador PID (Kp={Kp}, Ki={Ki}, Kd={Kd})", color='orange')
plt.axhline(y=1, color="r", linestyle=":", label="Referencia.") 

plt.title("Comparativa: Sistema Base vs. Sistema PID")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()

# --- Verificación de las métricas de desempeño ---
print("Métricas de desempeño del sistema con PID:")
info_PID = ct.step_info(sistema_PID, SettlingTimeThreshold=0.02)

print(f"  - Sobreimpulso (Overshoot): {info_PID['Overshoot']:.2f}%")
print(f"  - Tiempo de establecimiento (SettlingTime): {info_PID['SettlingTime']:.2f} s")
valor_final = y_pid[-1]
print(f"  - Valor final: {valor_final:.4f}")
print(f"  - Error en estado estacionario (1 - valor_final): {1 - valor_final:.4f}")
```
